```python
from tkinter import *
import pandas as pd

def submit_entries():
    
    """ Submit entries into dataframe then into excel
    
    """
    #steps:
    #create my file in the specified path
    #name the file
    #create labels in it for this example: (LABEL1,LABEL2,LABEL3)
    #execute code below
    
    file_path = 'TestingData.xlsx'  
    
    #ensure we dont overwrite
    df1 = pd.read_excel(file_path)
    
    #define series
    SeriesA = df1['LABEL1']
    SeriesB = df1['LABEL2']
    SeriesC = df1['LABEL3']
    
    #get operator values from entries
    for entry in entry1:
        A = pd.Series(entry.get())
        SeriesA = SeriesA.append(A)
  
    for entry in entrys2:
        B = pd.Series(entry.get())
        SeriesB = SeriesB.append(B)
       
    for entry in entrys3:
        C = pd.Series(entry.get())
        SeriesC = SeriesC.append(C)
 
    #pass back to data frame
    df2 = pd.DataFrame({"LABEL1":SeriesA, "LABEL2":SeriesB,'LABEL3':SeriesC })

    #read df2
    df2.to_excel(file_path, index=False)
    
master = Tk()

#make labeLs
w = Scale(master, orient=HORIZONTAL)
Label(master, text="LABEL1").grid(row=0, column=0)
Label(master, text="LABEL2").grid(row=0,column=1)
Label(master, text="LABEL3").grid(row=0,column=2)


entry1 = []
entrys2 =[]
entrys3 =[]
for i in range(10):
    col_entries1 = Entry(master)
    col_entries2 = Entry(master)
    col_entries3 = Entry(master)

    # position
    col_entries1.grid(row=i+1, column=0)
    col_entries2.grid(row=i+1, column=1)
    col_entries3.grid(row=i+1, column=2)

    entry1.append(col_entries1)
    entrys2.append(col_entries2)
    entrys3.append(col_entries3)

    
def handle_clipboard(event):
    """adds the paste feature to our tkinter gui"""
    
    #find index of current entry box in entry box Array
    try:
        index = entry1.index(event.widget)
        our_boxes = entry1
    except:
        try:
            index2 = entrys2.index(event.widget)
            index = index2
            our_boxes = entrys2
        except:
            index3 = entrys3.index(event.widget)
            index = index3
            our_boxes = entrys3
        else:
            index2 = entrys2.index(event.widget)
            index = index2
            our_boxes = entrys2
    else:
        index = entry1.index(event.widget)
        our_boxes = entry1
        
            
    #iterate form current location to length of selection
    lines = master.clipboard_get().split("\n")
    number_of_entrees = len(lines) - 1
    range_to_copy = our_boxes[index: index+number_of_entrees]
    
    #delete data in the entry boxes of interest
    for each_entry in range_to_copy:
        each_entry.delete(0, "end")
    
    #paste/instert data in the entry boxes of interest
    for each_entry, line in zip(range_to_copy, lines):
        each_entry.insert(0, line)
    
    return "break"

#executing the paste action
master.bind("<<Paste>>", handle_clipboard)

Button(master, text='Submit', command=submit_entries).grid(row=11, column=10)
menubar = Menu(master)
filemenu = Menu(menubar, tearoff=0)

menubar.add_cascade(label="File", menu=filemenu)
master.config(menu=menubar)


mainloop()
```
